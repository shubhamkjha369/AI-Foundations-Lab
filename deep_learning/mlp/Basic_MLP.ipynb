```python
import numpy as np

class MLP:

    def __init__(self, input_size, hidden_size, output_size, learning_rate = 0.1):
        self.lr = learning_rate

        #weights & biases
        self.W1 = np.random.randn(input_size, hidden_size)
        self.b1 = np.zeros(1, hidden_size)

        self.W2 = np.random.randn(hidden_size, output_size)
        self.b2 = np.zeros(1, output_size)

    # ACTIVATION FUNCTION
    def sigmoid(self, x):
        return 1/(1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1-x)

    # FORWARD PASS
    def forward(self, X):

        self.hidden_input = np.dot(X, self.W1) + self.b1
        self.hidden_ouput = self.sigmoid(self.hidden_input)

        self.final_input = np.dot(self.hidden_ouput, self.W2) + self.b2
        self.y_hat = self.sigmoid(self.final_input)

        return self.y_hat

    # BACKPROPOGATION
    
```
