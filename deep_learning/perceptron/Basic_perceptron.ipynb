A perceptron is the simplest neural network: one neuron that learns a linear decision boundary.

## Step 1: Understand the math

For inputs (x_1, x_2, ..., x_n): 

$$
y =
\begin{cases}
1, & \text{if } \sum_i w_i x_i + b > 0 \\
0, & \text{otherwise}
\end{cases}
$$

Where:

* `w` = weights
* `b` = bias
* `x` = input features
* `y` = predicted class (0 or 1)

### Perceptron Learning Rule

The weights and bias are updated according to:

$$
\mathbf{w} \leftarrow \mathbf{w} + \eta (y - \hat{y})\mathbf{x}
$$

$$
b \leftarrow b + \eta (y - \hat{y})
$$
where:

- $\eta$ = learning rate
- $y$ = true label
- $\hat{y}$ = perceptron's output
- $x$ = input vector
- $w$ = weight vector
- $b$ = bias term

---

## What happens during training?

For each sample:

1. Compute weighted sum.
2. Apply step activation.
3. Compare prediction with true label.
4. Compute error.
5. Update weights and bias.
6. Repeat until errors become zero.

## Limitation

A single perceptron can only solve **linearly separable** problems:

✅ AND
✅ OR
✅ NAND

❌ XOR

To solve XOR you need:

* Multiple perceptrons
* Hidden layers
* Backpropagation

That is the transition from a simple perceptron to a Multi-Layer Perceptron (MLP).

# Step 2: Perceptron from scratch

```python
import random

class perceptron:
    def __init__(self, n_inputs, learning_rate = 0.1):
        self.weights =  [random.uniform(-1, 1) for n in range(n_inputs)]
        self.bias = random.uniform(-1, 1)
        self.lr = learning_rate

    def activation(self, x):
        return 1 if x > 0 else 0

    def predict(self, inputs):
        total = self.bias

        for w, x in zip(self.weights, inputs):
            total += w * x

        return self.activation(total)

    def train(self, X, y, epochs = 10):

        for epoch in range(epochs):

            total_errors = 0

            for inputs, target in zip(X, y):

                prediction = self.predict(inputs)

                error = target - prediction

                total_errors += abs(error)

                for i in range(len(self.weights)):
                    self.weights[i] += self.lr * error * inputs[i]

                self.bias += self.lr * error

            print(f"Epoch: {epoch+1:2d} | Errors: {total_errors}")
```
