# Reshaping Tensors

```python
import torch
a = torch.ones(4, 4)
a
```

```python
# Reshape
a.reshape(2,2,2,2)
```

```python
# Flatten
a.flatten()
```

```python
# Permute
b = torch.rand(2, 3, 4)
b
```

```python
b.permute(2, 0, 1)
```

```python
# Unsqueeze
c = torch.rand(226, 226, 3)
c.unsqueeze(0).shape
```

```python
# Squeeze
d = torch.rand(1, 20)
d.squeeze(0).shape
```
