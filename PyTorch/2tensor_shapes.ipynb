# Tensor Shapes

```python
import torch
```

```python
x = torch.tensor([[1, 2, 3], [2, 4, 6]])
x
```

```python
x.shape
```

```python
# empty like - making tensor with same shape as existing tensor
torch.empty_like(x)
```

```python
torch.zeros_like(x)
```

```python
torch.ones_like(x)
```

```python
x.dtype
```

```python
# assign data type
a = torch.tensor([1.0, 2.0, 3.0], dtype = torch.int32)
a
```

```python
a.dtype
```

```python
b = torch.tensor([1.0, 2.0, 3.0], dtype = torch.float32)
b
```

```python
# find data type 
b.dtype
```

```python
torch.rand_like(b)
```

```python
torch.rand_like(a)
```

```python
# update datatype
x = x.to(torch.float32)
```

```python
torch.rand_like(x)
```
