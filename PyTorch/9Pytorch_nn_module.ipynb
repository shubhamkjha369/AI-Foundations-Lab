```python
import torch
import torch.nn as nn
```

```python
class Model1(nn.Module):

    def __init__(self, n_features):
        super().__init__()
        self.linear = nn.Linear(n_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        inp = self.linear(features)
        out = self.sigmoid(inp)
        return out
```

```python
features = torch.rand(10, 5)

model1 = Model1(features.shape[1])

model1(features)
```

```python
# Show model weights & bias
print("Weights --->", model1.linear.weight)
print("\nBias --->", model1.linear.bias)
```

```python
from torchinfo import summary
```

```python
summary(model1, input_size=(10, 5))
```

```python
import torch
import torch.nn as nn

class Model2(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        self.linear1 = nn.Linear(n_features, 5)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(5, 1)
        self.sigmoid = nn.Sigmoid()
    def forward(self, features):
        inp = self.linear1(features)
        out = self.relu(inp)
        inp = self.linear2(features)
        out = self.sigmoid(inp)
        return out
```

```python
features = torch.rand(10, 5)
model2 = Model2(features.shape[1])
model2(features)
```

```python
model2.linear1.weight
```

```python
summary(model2, input_size = (10, 5))
```

# Using Sequencial Container

```python
import torch
import torch.nn as nn
```

```python
class Model3(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(n_features, 5),
            nn.ReLU(),
            nn.Linear(5, 1),
            nn.Sigmoid(),
        )
    def forward(self, features):
        out = self.network(features)
        return out
```

```python
model3 = Model3(features.shape[1])
model3(features)
```
