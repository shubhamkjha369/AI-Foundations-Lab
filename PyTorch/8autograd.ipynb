## Disable Gradient Tracking

#### Option 1: `requires_grad = False`

```python
x = torch.tensor(3.0, requires_grad =True)
```

```python
x.requires_grad_(False)
x
```

```python
y = x ** 2
y
```

```python
y.backward()
```

#### Option 2: `detach()`

```python
x = torch.tensor(12.0, requires_grad =True)
```

```python
z=x.detach()
z
```

```python
y = x**2
y
```

```python
y1 = z**2
y1
```

```python
y.backward()
```

```python
x.grad
```

```python
x.grad.zero_()
```

```python
y1.backward()
```

#### Option 3: `torch.no_grad()`

```python
x = torch.tensor(6.0, requires_grad=True)
x
```

```python
with torch.no_grad():
    y = x**2
```

```python
y
```

```python
y.backward()
```
