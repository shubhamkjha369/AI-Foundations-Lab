## 3. Reduction Operation

```python
d = torch.randint(size = (2, 3), low=0, high=10)
d
```

```python
# sum
torch.sum(d)
```

```python
# sum along columns
torch.sum(d, dim=0)
```

```python
# sum along rows
torch.sum(d, dim = 1)
```

```python
# mean
print(d.dtype)
print(torch.mean(d, dtype = torch.float32))

# mean along column
print(torch.mean(d, dtype = torch.float32, dim=0))

# mean along row
print(torch.mean(d, dtype = torch.float32, dim=1))
```

```python
# median
torch.median(d)
```

```python
# min and max
print(torch.min(d))
print(torch.max(d))
```

```python
# product
torch.prod(d)
```

```python
# standard deviation
d= d.to(torch.float32)
torch.std(d)
```

```python
# Variance
torch.var(d)
```

```python
# argmax
torch.argmax(d)
```

```python
# argmin
torch.argmin(d)
```

## 4. Matrix Operations

```python
f = torch.randint(size=(2, 4), low=0, high=10)
g = torch.randint(size=(4, 1), low=0, high=10)

print(f)
print(g)
```

```python
# Matrix Multiplication
torch.matmul(f, g)
```

```python
# Dot Product
v1 = torch.tensor([1, 2])
v2 = torch.tensor([3, 4])

print(v1)
print(v2)
```

```python
torch.dot(v1, v2)
```

```python
# Transpose
print(f)
```

```python
torch.transpose(f, 0, 1)
```

```python
# Determinant
h = torch.randint(size=(2, 2), low=0, high=10, dtype=torch.float32)
h
```

```python
torch.det(h)
```

```python
# Inverse
torch.inverse(h)
```

## 5. Comparison Operations

```python
i = torch.randint(size=(2, 4), low=0, high=10)
j = torch.randint(size=(2, 4), low=0, high=10)

print(i)
print(j)
```

```python
# Greater than
i > j
```

```python
# less than
i < j
```

```python
# equal to
i == j
```

```python
# not equal to
i != j
```

```python
# greater than equal to
i >= j
```

```python
# less than equal to
i <= j
```

## 6. Special Functions

```python
k = torch.randint(size=(2, 3), low=0, high=10)
k
```

```python
# log
torch.log(k)
```

```python
# exponents
torch.exp(k)
```

```python
# Sqrt
torch.sqrt(k)
```

```python
# Sigmoid
torch.sigmoid(k)
```

```python
# Softmax
k = k.to(torch.float32)
torch.softmax(k, dim=0)
```

```python
# ReLu
torch.relu(k)
```
